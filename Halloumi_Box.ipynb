Halloumi Box

```python
t = int(input())
for _ in range(t):
    n, k = map(int, input().split())
    arr = list(map(int, input().split()))

    ok = True
    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            ok = False
            break

    if ok == True or k > 1:
        print("YES")
    else:
        print("NO")
```
